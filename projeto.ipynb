```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

num_vertices = 15
probabilidade_arestas = 0.1
arquivo_saida = "grafo.txt"

grafo = nx.Graph()
grafo.add_nodes_from(range(num_vertices))

# cria arestas
for i in range(num_vertices):
    for j in range(i + 1, num_vertices):
        if random.random() < probabilidade_arestas:
            peso = random.randint(1, 10)
            grafo.add_edge(i, j, weight=peso)

# exporta grafo.txt
with open(arquivo_saida, "w") as f:
    for u, v, peso in grafo.edges(data="weight"):
        f.write(f"{u} {v} {peso}\n")

print(f"Grafo exportado para {arquivo_saida}")

# matriz de adjacencia
matriz_adjacencia = nx.adjacency_matrix(grafo, weight="weight").todense()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 20))

# plot
pos = nx.circular_layout(grafo)
nx.draw(grafo, pos, with_labels=True, node_size=300, font_size=16, node_color="skyblue", edge_color="gray", ax=ax1)
edge_labels = nx.get_edge_attributes(grafo, 'weight')
nx.draw_networkx_edge_labels(grafo, pos, edge_labels=edge_labels, font_size=14, ax=ax1)
ax1.set_title("Grafo com 100 vértices, arestas aleatórias e pesos")

# plot da matriz de adjacência
ax2.imshow(matriz_adjacencia, cmap="Blues", interpolation="none")
for i in range(num_vertices):
    for j in range(num_vertices):
        value = matriz_adjacencia[i, j]
        if value > 0:
            ax2.text(j, i, int(value), ha='center', va='center', color="black", fontsize=16)

ax2.set_title("Matriz de Adjacência com Pesos")
ax2.set_xlabel("Vértices")
ax2.set_ylabel("Vértices")

plt.tight_layout()
plt.show()
```
